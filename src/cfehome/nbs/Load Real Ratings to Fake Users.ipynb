```python
from django.contrib.contenttypes.models import ContentType
ctype = ContentType.objects.get(app_label="movies", model="movie")
```

```python
from django.db import transaction

def chunk_list(lst, chunk_size):
    for i in range(0, len(lst), chunk_size):
        yield lst[i:i + chunk_size]
        

def bulk_create_ratings(rating_records, ctype, batch_size=1000):
    new_ratings = []
    for chunk in chunk_list(rating_records, batch_size):
        for r in chunk:
            r['content_type'] = ctype
            new_ratings.append(
                Rating(**r)
            )
        with transaction.atomic():
            Rating.objects.bulk_create(new_ratings, ignore_conflicts=True)
            new_ratings = []
            
bulk_create_ratings(rating_records, ctype)
```
